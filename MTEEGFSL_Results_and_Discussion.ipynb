```python
# define the data
x = ['EEGNet', 'HTNet', 'Attention based FSL', 'Transfer Learning based FSL', 'Gradual Relational Network', 'Reptile based FSL', 'Ours']
y = [64.00, 72.00, 73.60, 76.30, 80.85, 82.24, 88.30]

# create the bar plot
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b', '#e377c2']
fig, ax = plt.subplots(figsize=(8, 6))
bars = ax.bar(x, y, color=colors)

# customize the appearance
plt.xlabel('Methods')
plt.ylabel('Accuracy')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--')

# add data labels inside bars
for i, v in enumerate(y):
    ax.text(i, v+1, str(v), ha='center', fontsize=10)
    bars[i].set_zorder(0)

# add average line
plt.axhline(y=75, color='gray', linestyle='--')

# add legend
plt.legend(['Average'], loc='upper left', fontsize=8)

# adjust the plot
plt.subplots_adjust(bottom=0.25)

# rotate x-axis labels vertically
for tick in ax.get_xticklabels():
    tick.set_rotation(90)

# save figure to file
plt.savefig('Comparison_results.png', dpi=300)

# display the plot
plt.show()
```
